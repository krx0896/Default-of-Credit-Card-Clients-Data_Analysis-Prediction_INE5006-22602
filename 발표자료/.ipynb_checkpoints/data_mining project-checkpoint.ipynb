## 상관관계 큰 변수끼리 묶어서 파생 변수 만들기

```python
df['pay_total'] = df['PAY_0'] + df['PAY_2'] + df['PAY_3'] + df['PAY_4'] + df['PAY_5'] + df['PAY_6']
```

```python
df['bill_amt_total'] = df['BILL_AMT1'] + df['BILL_AMT2'] + df['BILL_AMT3'] + df['BILL_AMT4'] + df['BILL_AMT5'] + df['BILL_AMT6']
```

```python
df['pay_amt_total'] = df['PAY_AMT1'] + df['PAY_AMT2'] + df['PAY_AMT3'] + df['PAY_AMT4'] + df['PAY_AMT5'] + df['PAY_AMT6']
```

```python
df.head()
```

## 나머지 데이터 지우기

```python
df2 = df.drop([df.columns[5], df.columns[6], df.columns[7], df.columns[8], df.columns[9], df.columns[10], df.columns[11], df.columns[12], df.columns[13], df.columns[14], df.columns[15], df.columns[16], df.columns[17], df.columns[18], df.columns[19], df.columns[20], df.columns[21], df.columns[22]],  axis=1)
```

```python
df2.head()
```

## 카드 연체 이력 변수 오브젝트로 타입 재지정

```python
df2 = df2.astype({'default payment next month': 'object'})
```
