```python
col_candidate = []

#COl 기준 데이터의 수가 700개 이상인 것만 살리기
for idx, val in zip(df.columns, df.describe().iloc[0]):
  if val >= 700:
    col_candidate.append(idx)
df = df[col_candidate]
print(len(col_candidate))


col_candidate = []
#COl 기준 std가 0이면 제외하기
for idx, val in zip(df.columns, df.describe().iloc[2]):
  if val != 0:
    col_candidate.append(idx)
df = df[col_candidate]
print(len(col_candidate))
```

## 10-fold

```python
kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 0)
```

## Roburst

```python
transformer = RobustScaler()
```

### 4개의 모델 정의

```python
model1 = LogisticRegression()
model2 = RandomForestClassifier()
model3 = AdaBoostClassifier(n_estimators=50, learning_rate=0.01)
model4 = SVC(kernel = 'linear')
```

### 로지스틱회귀 성능지표

```python
accuracy1 = []
precision1 = []
recall1 = []
f11 = []
for idx, (train_idx, test_idx) in enumerate(kfold.split(X, y)):
    train = X.loc[train_idx]
    test = X.loc[test_idx]
    
    X_train  = transformer.fit_transform(train)
    X_test = transformer.transform(test)
    
    X_resampled, y_resampled = SMOTE(random_state=0).fit_resample(X_train, y.loc[train_idx])
    
    model1.fit(X_resampled, y_resampled)
    Logistic_pred = model1.predict(X_test)
    
    accuracy1.append(accuracy_score(y.loc[test_idx],Logistic_pred))
    precision1.append(precision_score(y.loc[test_idx],Logistic_pred))
    recall1.append(recall_score(y.loc[test_idx],Logistic_pred))
    f11.append(f1_score(y.loc[test_idx],Logistic_pred))
```

```python
np.array(accuracy1).mean()
```
